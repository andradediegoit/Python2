```python
class Aluno:
    def __init__(self, nome, numero, idade, turma):
        self.nome = nome
        self.numero = numero
        self.idade = idade
        self.turma = turma
    
    def obter_nome(self):
        return self.nome
    
    def obter_numero(self):
        return self.numero
    
    def obter_idade(self):
        return self.idade
    
    def obter_turma(self):
        return self.turma

def criar_aluno():
    nome = input("Digite o nome do aluno: ")
    numero = input("Digite o número do aluno: ")
    idade = input("Digite a idade do aluno: ")
    turma = input("Digite a turma do aluno: ")
    return Aluno(nome, numero, idade, turma)

def listar_alunos(alunos):
    print("Alunos existentes:")
    for aluno in alunos:
        print(f"Nome: {aluno.obter_nome()}, Número: {aluno.obter_numero()}, Idade: {aluno.obter_idade()}, Turma: {aluno.obter_turma()}")

def main():
    alunos = []
    while True:
        print("\n1. Criar novo aluno")
        print("2. Consultar alunos existentes")
        print("3. Sair")
        opcao = input("Selecione uma opção: ")

        if opcao == "1":
            novo_aluno = criar_aluno()
            alunos.append(novo_aluno)
            print("Aluno criado com sucesso!")
        elif opcao == "2":
            if not alunos:
                print("Não existem alunos cadastrados.")
            else:
                listar_alunos(alunos)
        elif opcao == "3":
            print("Programa encerrado.")
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
```
