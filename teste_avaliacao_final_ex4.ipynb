```python
import random
import csv
import matplotlib.pyplot as plt


def create_key():
    numbers = random.sample(range(1, 51), 5)  # Generates a sequence of 5 numbers from 1 to 50 without repetition
    stars = random.sample(range(1, 13), 2)    # Generates a sequence of 2 numbers from 1 to 12 without repetition 
    return numbers + stars                    # return the keys as a tuple containing the two sequences

def create_keys(num_keys):                    # function to generete multiple keys
    keys = []
    for _ in range(num_keys):
        key = create_key()
        keys.append(key)
    return keys

keys = create_keys(50)    # Generate 30 unique keys


 # Create a CSV file to store the keys
with open('Keys.csv', 'w', newline='') as cvsfile:           
    write = csv.writer(cvsfile)
    write.writerow(['Number 1','Number 2', 'Number 3','Number 4', 'Number 5','Star 1', 'Star 2'])
    write.writerows(keys)


# grafic

numbers = [number for key in keys for number in key[:5]]
plt.figure(figsize=(12, 4))
plt.hist(numbers, bins=range(1, 52), alpha=0.50, label='Numbers', color='darkblue')
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.title('Frequency of Generated Numbers')
plt.legend()
plt.grid(False)
plt.show()


stars = [star for key in keys for star in key[5:]]
plt.figure(figsize=(12, 4))
plt.hist(stars, bins=range(1,12), alpha=0.25, label='Stars', color='orange')
plt.xlabel('Star')
plt.ylabel('Frequency')
plt.title('Frequency of Generated Stars')
plt.legend()
plt.grid(False)
plt.show()
```

```python
class Aluno:
    def __init__(self, nome, numero, idade, turma):
        self.nome = nome
        self.numero = numero
        self.idade = idade
        self.turma = turma
    
    def obter_nome(self):
        return self.nome
    
    def obter_numero(self):
        return self.numero
    
    def obter_idade(self):
        return self.idade
    
    def obter_turma(self):
        return self.turma

# Função para criar um novo aluno
def criar_aluno():
    nome = input("Digite o nome do aluno: ")
    numero = input("Digite o número do aluno: ")
    idade = input("Digite a idade do aluno: ")
    turma = input("Digite a turma do aluno: ")
    return Aluno(nome, numero, idade, turma)

# Função para listar todos os alunos existentes
def listar_alunos(alunos):
    print("Alunos existentes:")
    for aluno in alunos:
        print(f"Nome: {aluno.obter_nome()}, Número: {aluno.obter_numero()}, Idade: {aluno.obter_idade()}, Turma: {aluno.obter_turma()}")

# Função principal
def main():
    alunos = []
    while True:
        print("\n1. Criar novo aluno")
        print("2. Consultar alunos existentes")
        print("3. Sair")
        opcao = input("Selecione uma opção: ")

        if opcao == "1":
            novo_aluno = criar_aluno()
            alunos.append(novo_aluno)
            print("Aluno criado com sucesso!")
        elif opcao == "2":
            if not alunos:
                print("Não existem alunos cadastrados.")
            else:
                listar_alunos(alunos)
        elif opcao == "3":
            print("Programa encerrado.")
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
```
