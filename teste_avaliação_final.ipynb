```python
import random
import csv
import matplotlib.pyplot as plt

# Função para gerar uma chave única
def gerar_chave():
    # Gera uma sequência de 3 números de 1 a 30 sem repetição
    numeros = random.sample(range(1, 31), 3)
    # Gera uma sequência de 2 números de 1 a 10 sem repetição
    estrelas = random.sample(range(1, 11), 2)
    # Retorna a chave como uma tupla contendo as duas sequências
    return numeros + estrelas

# Função para gerar múltiplas chaves
def gerar_chaves(num_chaves):
    chaves = []
    for _ in range(num_chaves):
        chave = gerar_chave()
        chaves.append(chave)
    return chaves

# Gera 30 chaves únicas
chaves = gerar_chaves(30)

# Salva as chaves em um arquivo CSV
with open('chaves.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['num1', 'num2', 'num3', 'estrela1', 'estrela2'])
    writer.writerows(chaves)

# Preparando dados para o gráfico
numeros = [numero for chave in chaves for numero in chave[:3]]
estrelas = [estrela for chave in chaves for estrela in chave[3:]]

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.hist(numeros, bins=range(1, 32), alpha=0.5, label='Números')
plt.hist(estrelas, bins=range(1, 12), alpha=0.5, label='Estrelas')
plt.xlabel('Número/Estrela')
plt.ylabel('Frequência')
plt.title('Frequência de Números e Estrelas Gerados')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import random
import matplotlib.pyplot as plt

def create_key():
    numbers = random.sample(range(1, 51), 5)  # Gera uma sequência de 5 números de 1 a 50 sem repetição
    stars = random.sample(range(1, 13), 2)    # Gera uma sequência de 2 números de 1 a 12 sem repetição 
    return numbers + stars                    # Retorna as chaves como uma tupla contendo as duas sequências

def create_keys(num_keys):                    # Função para gerar múltiplas chaves
    keys = []
    for _ in range(num_keys):
        key = create_key()
        keys.append(key)
    return keys

keys = create_keys(50)                        # Gera 50 chaves únicas

# Contagem da frequência dos números e estrelas
frequency_numbers = [0] * 50
frequency_stars = [0] * 12

for key in keys:
    for number in key[:5]:
        frequency_numbers[number - 1] += 1  # Decrementa 1 para corresponder aos índices da lista
    for star in key[5:]:
        frequency_stars[star - 1] += 1      # Decrementa 1 para corresponder aos índices da lista

# Preparando os dados para o gráfico
x_numbers = range(1, 51)
x_stars = range(1, 13)

# Plotando o gráfico
bar_width = 0.70
plt.figure(figsize=(12, 4), facecolor='lightgray')
plt.bar(x_numbers, frequency_numbers, color='darkblue' , width = bar_width, alpha=0.7, label='Numbers')
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.title('Frequency of Numbers')
plt.legend()
plt.tight_layout()
plt.grid(False)

for x, y in zip(x_numbers, frequency_numbers):
    plt.text(x, y + 0.05, str(y), ha='center', va='bottom')

plt.show()

bar_width = 0.70
plt.figure(figsize=(12, 5), facecolor=  'lightgray')
plt.bar(x_stars, frequency_stars, color='darkorange', width = bar_width, alpha=0.7, label='Stars')
plt.xlabel('Star')
plt.ylabel('Frequency')
plt.title('Frequency of Stars')
plt.legend()
plt.tight_layout()
plt.grid(False)

for x, y in zip(x_stars, frequency_stars):
    plt.text(x, y + 0.05, str(y), ha='center', va='bottom')

plt.show()
```
