```python
import random
import csv
import matplotlib.pyplot as plt

# Função para gerar uma chave única
def gerar_chave():
    # Gera uma sequência de 3 números de 1 a 30 sem repetição
    numeros = random.sample(range(1, 31), 3)
    # Gera uma sequência de 2 números de 1 a 10 sem repetição
    estrelas = random.sample(range(1, 11), 2)
    # Retorna a chave como uma tupla contendo as duas sequências
    return numeros + estrelas

# Função para gerar múltiplas chaves
def gerar_chaves(num_chaves):
    chaves = []
    for _ in range(num_chaves):
        chave = gerar_chave()
        chaves.append(chave)
    return chaves

# Gera 30 chaves únicas
chaves = gerar_chaves(30)

# Salva as chaves em um arquivo CSV
with open('chaves.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['num1', 'num2', 'num3', 'estrela1', 'estrela2'])
    writer.writerows(chaves)

# Preparando dados para o gráfico
numeros = [numero for chave in chaves for numero in chave[:3]]
estrelas = [estrela for chave in chaves for estrela in chave[3:]]

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.hist(numeros, bins=range(1, 32), alpha=0.5, label='Números')
plt.hist(estrelas, bins=range(1, 12), alpha=0.5, label='Estrelas')
plt.xlabel('Número/Estrela')
plt.ylabel('Frequência')
plt.title('Frequência de Números e Estrelas Gerados')
plt.legend()
plt.grid(True)
plt.show()

```
