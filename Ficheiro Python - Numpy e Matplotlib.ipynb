## 2.1. Desenhar gráficos com função linspace

```python
import numpy as np
import matplotlib.pyplot as plt # Importar o módulo especifico pyplot do matplotlib
```

```python
import numpy as np
from matplotlib import pyplot as plt # Importar o módulo especifico pyplot do matplotlib

array_user = np.linspace(0,5,100)

# array_yser -> eixo x
# array_user -> eixo y

plt.plot(array_user,array_user, label = 'linear') # Legenda especificada na função plot

plt.xlabel('Eixo X') # Etiqueta eixo X
plt.ylabel('Eixo Y') # Etiqueta eixo Y
plt.title('Gráfico simples') # Colocar titulo no gráfico
plt.legend() # Colocar legenda no gráfico
plt.savefig('exp.png') # Guardar a imagem na mesma pasta do ficheiro Python a executar
plt.show() # Mostar o gráfico na janela
```

## Gráfico como multiplas caracteristicas

```python
import numpy as np
from matplotlib import pyplot as plt # Importar o módulo especifico pyplot do matplotlib

array_user = np.linspace(0,2,1000)

plt.plot(array_user, array_user)
plt.plot(array_user, array_user ** 2)

# array1 = [1 2 3 4 5 6 7 8] - Eixo X
# array2 = [1 4 9 16 25 36 49 64] - Eixo Y

plt.plot(array_user, array_user ** 3)

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico multiplas características')
plt.legend(['linear','quadrado','cubo'])
plt.show()

```

## 2.2. Desenhar gráficos com listas e opções
### 2.2.1. Gráfico com linha e marcador

```python
from matplotlib import pyplot as plt # Importar o módulo especifico pyplot do matplotlib

plt.plot([1,2,3,4],[1,4,9,16], color = 'orange', marker = 'o') # Colocar cor personalizada e marcador de pontos de interseção
plt.axis([0,6,0,20]) # Definição do eixo X e Y [ponto1_x,ponto2_x,ponto1_y,ponto2_y]
plt.show()
```

### 2.2.2. Gráficos com linha e marcador por abreviação

```python
from matplotlib import pyplot as plt # Importar o módulo especifico pyplot do matplotlib

plt.plot([1,2,3,4],[1,4,9,16], 'b:X') # Colocar cor personalizada e marcador de pontos de interseção
plt.axis([0,6,0,20]) # Definição do eixo X e Y [ponto1_x,ponto2_x,ponto1_y,ponto2_y]

plt.show()

'''
'b' : blue
'g' : green
'r' : red
'c' : cyan
'm' : magenta
'y' : yellow
'k' : black
'w' : white

'-' : linha continua
'--' : linha tracejada
'-.' : tracejado com ponto
':' : linha com pontos

'o' : marcador em circulo
's' : marcador em quadrado
'^' : marcador em triângulo
'X' : marcador com X

'''
```

### 2.2.3. Gráfico sem linhas de marcador por parametros de opção

### 2.2.4. Gráfico com multiplas linhas com marcadores por abreviatura

```python
import numpy as np
from matplotlib import pyplot as plt # Importar o módulo especifico pyplot do matplotlib

array_user = np.arange(0,5,0.2) # arange: Cria um arranjo de numeros semelhante ao range, mas permite numeros do tipo float

plt.plot(array_user, array_user, 'r--')
plt.plot(array_user, array_user ** 2, 'bs')

# array1 = [1 2 3 4 5 6 7 8] - Eixo X
# array2 = [1 4 9 16 25 36 49 64] - Eixo Y

plt.plot(array_user, array_user ** 3, 'g^')

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico multiplas características')
plt.legend(['linear','quadrado','cubo'])
plt.show()
```

## 2.3. Gráfico Polar

```python
import numpy as np
from matplotlib import pyplot as plt # Importar o módulo especifico pyplot do matplotlib

raio = np.arange(0,2,0.01)

perimetro = 2 * np.pi * raio

polar_plt = plt.subplot(111, projection = 'polar') #111: 1ª Linha, 1ª Coluna, 1ª Posição
polar_plt.plot(perimetro, raio)
polar_plt.set_rticks([0.5,1,1.5,2]) # Especificar o numero de eixos dentro do gáfico polar
polar_plt.set_title('Gráfico Polar')
polar_plt.grid(True)
plt.show()
```

## 2.4. Vários gráficos no mesmo plot

```python
import numpy as np
from matplotlib import pyplot as plt # Importar o módulo especifico pyplot do matplotlib

array_user = np.linspace(0,2,50)

plt.subplot(1,2,1) # 1 - Plot com 1 linha; 2 - Plot com 2 colunas; 1 - Plot da posição 1 a ser declarado

plt.plot(array_user, array_user, color = 'red', linestyle = '-.')
plt.plot(array_user, array_user ** 2, 'bs')
plt.plot(array_user, array_user ** 2, 'g^')

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de linhas')
plt.legend(['linear','quadrado','cubo'])

plt.subplot(1,2,2) # 1- Plot com 1 linha; 2 - Plot com 2 Colunas; 2 - Plot da posição 2 a ser declarado

array_num = np.arange(0,10)

arrayRnd = np.random.randint(0,100,10)
arr

plt.bar(array_num,arrayRnd, color = 'blue')

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de barras')
plt.legend(['numeros aleatórios'])
plt.tight_layout() # Agrupar gráficos de forma a não se sobreporem
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt # Importar o módulo especifico pyplot do matplotlib

array_user = np.linspace(0,2,50)

plt.subplot(2,1,1) # 1 - Plot com 2 linhas; 2 - Plot com 1 coluna; 1 - Plot da posição 1 a ser declarado

plt.plot(array_user, array_user, color = 'red', linestyle = '-.')
plt.plot(array_user, array_user ** 2, 'bs')
plt.plot(array_user, array_user ** 2, 'g^')

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de linhas')
plt.legend(['linear','quadrado','cubo'])

plt.subplot(2,1,2) # 1- Plot com 2 linhas; 2 - Plot com 1 Coluna; 2 - Plot da posição 2 a ser declarado

array_num = np.arange(0,10)

arrayRnd = np.random.randint(0,100,10)
arrayRnd2 = np.random.randint(0,100,10)

plt.bar(array_num,arrayRnd)
plt.bar(array_num,arrayRnd2)

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de barras')
plt.legend(['numeros aleatórios'])
plt.tight_layout() # Agrupar gráficos de forma a não se sobreporem
plt.show()
```
